## 1- Import Necessary Libraries

In this step, we import the libraries required for our analysis. We'll be using pandas for data manipulation, matplotlib for plotting, numpy for numerical computations, and networkx for working with graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx

from TwitterGraphSimulator import TwitterGraphSimulator
```

## 2- Load Twitter Data

In this step, we load the Twitter data from the CSV file `canada_tweets.csv` into a pandas DataFrame. We handle bad lines by skipping them, and we specify the quote and escape characters accordingly.

```python
df_tweets = pd.read_csv('../data/canada_tweets.csv', on_bad_lines='skip', quotechar="'", escapechar='\\')
df_tweets.head()
```

## 3- Simulate Twitter-Like Graph

In this step, we set the parameters for simulating a Twitter-like graph and use the `TwitterGraphSimulator` class to perform the simulation.

```python
AVERAGE_NUM_FOLLOWING = 50
MAX_ITERATIONS = 1

names = df_tweets[' screen_name']

simulator = TwitterGraphSimulator(AVERAGE_NUM_FOLLOWING, MAX_ITERATIONS)
results = simulator.simulate(names)
results
```

## 4- Analysis

In this step, we extract the results from the simulation and analyze them.

```python
nodes, edges, iterations, lengths_nodes, lengths_edges = results
print('Number of Nodes:', len(nodes))
print('Number of Edges:', len(edges))
```

## 5- Visualization

In this step, we visualize the simulated Twitter-like graph using NetworkX and Matplotlib.

```python
G = nx.Graph()

G.add_nodes_from(nodes)

G.add_edges_from(edges)

pos = nx.spring_layout(G, k=0.1)

nx.draw(G, with_labels=True, font_size=7, edge_color='gray', pos=pos)
plt.title("Twitter Follow Graph - Small")
plt.savefig('../graphs/small_follow_graph.png')
plt.show()
```

## 6- Results and Conclusions

### Simulation Results:
- **Number of Nodes:** The simulation resulted in a total of 50 nodes representing Twitter users.
- **Number of Edges:** The graph contains 49 edges representing the connections (follow relationships) between users.
- **Graph Structure:** The Twitter-like graph exhibits a tree structure.

### Observations:
- **Node Distribution:** The distribution of followers among users follows a normal distribution with an average of 700 which was reduced to 50 for simulation purposes.
